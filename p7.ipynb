```python
text = """Tokenization is the first step in text analytics. 
The process of breaking down a text paragraph into smaller chunks 
such as words or sentences is called Tokenization."""
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
sentences = sent_tokenize(text)
print("Sentence Tokenization:\n", sentences)
```

```python
words = word_tokenize(text)
print("\nWord Tokenization:\n", words)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
```

```python

clean_text = re.sub('[^a-zA-Z]', ' ', text)
```

```python
tokens = word_tokenize(clean_text.lower())
```

```python
filtered_words = [w for w in tokens if w not in stop_words]
print("\nFiltered Words:\n", filtered_words)
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
stemmed_words = [ps.stem(w) for w in filtered_words]
print("\nStemmed Words:\n", stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(w) for w in filtered_words]
print("\nLemmatized Words:\n", lemmatized_words)
nltk.download('averaged_perceptron_tagger_eng')
```

```python
pos_tags = nltk.pos_tag(filtered_words)
print("\nPOS Tagging:\n", pos_tags)
```
