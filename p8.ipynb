```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
pt=[20,25,28,30,36,40,43]
pt2=[21,32,36,35,39,41,45]
pt3=[20,23,25,28,32,33]
```

```python
plt.plot(pt)
plt.plot(pt2)
plt.plot(pt3)
plt.show()
```

```python

plt.plot(pt, c='r', marker='o', mec='g', mfc='y')
plt.plot(pt2, c='b', marker='X', mec='g', mfc='y')
plt.xlabel('sample point')
plt.ylabel('value')
plt.title('sample vs value')
plt.legend(['DBMS','CN'])
plt.show()
```

```python

plt.plot(pt, c='r', marker='o', ls='--', mec='g', mfc='y')
plt.plot(pt2, c='b', marker='X', ls=':', mec='k', mfc='C6')
plt.plot(pt3, c='k', marker='s', ls="-.", mec='c', mfc='m')
plt.xlabel('sample point')
plt.ylabel('value')
plt.title('sample vs value')
plt.legend(['DBMS','CN','DSA'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = sns.load_dataset("titanic")
```

```python
dataset.head()
```

```python

sns.distplot(x = dataset['age'], bins = 10) 
```

```python
sns.distplot(dataset['age'], bins = 10,kde=False)
```

```python
sns.jointplot(x = dataset['age'], y = dataset['fare'], kind = 'scatter') 
```

```python
sns.jointplot(x = dataset['age'], y = dataset['fare'], kind = 'hex')
```

```python

sns.rugplot(dataset['fare']) 
```

```python

sns.barplot(x='sex', y='age', data=dataset, hue='sex')
```

```python
sns.barplot(x='sex', y='age', data=dataset, estimator=np.std,hue='sex') 
```

```python
sns.countplot(x='sex', data=dataset,hue='sex')
```

```python
sns.boxplot(x='sex', y='age', data=dataset,hue='sex') 
```

```python
sns.boxplot(x='sex', y='age', data=dataset, hue="survived")
```

```python

sns.violinplot(x='sex', y='age', data=dataset,hue='sex')
```
