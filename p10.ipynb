```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=sns.load_dataset("iris")
dataset.head()
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python

dataset.hist(figsize=(10,8))
plt.suptitle("Histogram of Iris Features")
plt.show()
```

```python

dataset.boxplot(figsize=(10,6))
plt.title("Boxplot of Iris Features")
plt.show()
```

Observations For Histogram :- Most values of sepal length are between 5 to 6 cm Sepal width values are mostly between 2.5 to 3.5 cm Petal length shows clear grouping (different species) Petal width also shows distinct clusters

Observation For Boxplot :- Sepal width has some outliers (very small or very large values) Other features (sepal length, petal length, petal width) have few or no outliers Petal features show clear variation between species Sepal features overlap more, so they are less useful for classification
